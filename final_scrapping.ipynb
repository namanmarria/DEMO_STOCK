```python
def scrap_insert_data(url):
    mycursor=mydb.cursor(buffered = True)
    dt=str(datetime.datetime.now())
    driver.get(url)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    fin_streamer = soup.find_all('div',{'class':'D(ib) Fw(200) Mend(25px) Mend(20px)--lgv3'})
    for i in fin_streamer:
        price=i.find('fin-streamer',{'class':'Fw(b) Fz(m) Mb(-4px) D(ib)'}).text
        changeamount=i.find('fin-streamer',{'class':'Fw(500) Pstart(10px) Fz(m)'}).text
        changepercentage=i.find('fin-streamer',{'data-field':'regularMarketChangePercent'}).text
    if 'CIM' in url:
        name="CIM"
    elif 'PFE' in url:
        name="PFIZER"
    elif 'BNB' in url:
        name="BNB-USD"
    elif 'MGM' in url:
        name="MGM"        
    elif 'JNJ' in url:
        name="Johnson-Johnson"
    elif 'ADA' in url:
        name="Cardano_ADA"
    
    #query to execute sql
    variables={'name':name,'price':price,'changeamount':changeamount,'changepercentage':changepercentage,'datetime':dt}
    sql="insert into stock values('{name}','{price}','{changeamount}','{changepercentage}','{datetime}')".format_map(variables)
    mycursor.execute(sql)
    mydb.commit()
    print("Data Inserted")
    return(name,'',price,'',changeamount,'',changepercentage,'',dt)   
```
